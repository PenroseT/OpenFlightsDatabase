```python
import numpy as np
import matplotlib.pyplot as plt
import openflights as of
```

```python
world = of.create_world()

route_functions = [of.haversine_distance]
route_params = ({"limit":30})
airport_functions = [of.haversine_distance]
airport_params = ({"limit":30, "rescale_factor":1e-3})

route_query = """SELECT r.source_airport_name AS src_nm, r.destination_airport_name AS dst_nm,
                  src.longitude AS src_lon, src.latitude AS src_lat,
                  dst.longitude AS dst_lon, dst.latitude AS dst_lat,
                  py_haversine_distance(src.longitude,  src.latitude,
                                        dst.longitude, dst.latitude) AS route_dist
                FROM routes r
                INNER JOIN airports src ON r.source_airport_id = src.airport_id
                INNER JOIN airports dst ON r.destination_airport_id = dst.airport_id
                ORDER BY route_dist DESC
                LIMIT :limit"""

airport_query = """SELECT r.source_airport_name AS airport_nm,
                          src.longitude AS airport_lon,
                          src.latitude AS airport_lat,
                          py_haversine_distance(src.longitude,  src.latitude,
                                                dst.longitude, dst.latitude) AS route_dist,
                          :rescale_factor*py_haversine_distance(src.longitude,  src.latitude,
                                                dst.longitude, dst.latitude) AS airport_size
                FROM routes r
                INNER JOIN airports src ON r.source_airport_id = src.airport_id
                INNER JOIN airports dst ON r.destination_airport_id = dst.airport_id
                ORDER BY route_dist DESC
                LIMIT :limit"""

gdf_routes = of.create_routes("openflights.db", route_query, route_functions, route_params)
gdf_airports = of.create_airports("openflights.db", airport_query, airport_functions, airport_params)
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,9))

# Constructs the choropleth world with colors representing
# the estimated populations of countries
of.populated_world(world, ax)

# Plot the line between the cities
gdf_routes.plot(ax=ax, linewidth=2.0, color='#515582', alpha=0.5)

# Plot the airports
gdf_airports.plot(ax=ax, alpha=0.5, color='#15198f',
                  markersize=gdf_airports['airport_sizes'])

plt.tick_params(axis="both", labelsize=20)
plt.xlabel('Longitude', fontsize=35)
plt.ylabel('Latitude', fontsize=35)

plt.show()
```

```python
world = of.create_world()

route_functions = [of.haversine_distance, of.route_identifier]
route_params = ({"limit":20})
airport_functions = [of.haversine_distance, of.route_identifier]
airport_params = ({"limit":80, "rescale_factor":5e-3})

route_query = """SELECT r.source_airport_name AS src_nm, r.destination_airport_name AS dst_nm,
                  src.longitude AS src_lon, src.latitude AS src_lat,
                  dst.longitude AS dst_lon, dst.latitude AS dst_lat,
                  py_haversine_distance(src.longitude,  src.latitude,
                                        dst.longitude, dst.latitude) AS route_dist
                FROM routes r
                    INNER JOIN airports src ON r.source_airport_id = src.airport_id
                    INNER JOIN airports dst ON r.destination_airport_id = dst.airport_id
                WHERE r.stops = 0
                GROUP BY py_route_identifier(r.source_airport_id, r.destination_airport_id)
                ORDER BY route_dist DESC
                    LIMIT :limit"""

airport_query = """SELECT r.source_airport_name AS airport_nm,
                          src.longitude AS airport_lon,
                          src.latitude AS airport_lat,
                          py_haversine_distance(src.longitude,  src.latitude,
                                                dst.longitude, dst.latitude) AS route_dist,
                          :rescale_factor*py_haversine_distance(src.longitude,  src.latitude,
                                                dst.longitude, dst.latitude) AS airport_size
                FROM routes r
                    INNER JOIN airports src ON r.source_airport_id = src.airport_id
                    INNER JOIN airports dst ON r.destination_airport_id = dst.airport_id
                WHERE r.stops = 0
                ORDER BY route_dist DESC
                    LIMIT :limit"""

gdf_routes = of.create_routes("openflights.db", route_query, route_functions, route_params)
gdf_airports = of.create_airports("openflights.db", airport_query, airport_functions, airport_params)
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,9))

# Constructs the choropleth world with colors representing
# the estimated populations of countries
of.populated_world(world, ax)

# Plot the line between the cities
gdf_routes.plot(ax=ax, linewidth=1.0, color='#515582', alpha=0.5)

# Plot the airports
gdf_airports.plot(ax=ax, alpha=0.5, color='#15198f',
                  markersize=gdf_airports['airport_sizes'])

plt.tick_params(axis="both", labelsize=20)
plt.xlabel('Longitude', fontsize=35)
plt.ylabel('Latitude', fontsize=35)
plt.show()
```
